````python
# -------------------------------------------------------------
# EDA for CSE-151B Climate Dataset (uses YOUR ClimateDataModule)
# -------------------------------------------------------------
import os
from pathlib import Path
from datetime import datetime
import numpy as np
import pandas as pd
import xarray as xr
import dask.array as da
import torch
from torch.utils.data import Dataset, DataLoader

import torch.nn as nn
import torch.optim as optim
import lightning.pytorch as pl

import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.dates as mdates
from matplotlib.gridspec import GridSpec
````

````python
class ClimateDataset(Dataset):
    def __init__(self, inputs_dask, outputs_dask, output_is_normalized=True):
        self.size = inputs_dask.shape[0]
        print(f"Creating dataset with {self.size} samples...")

        inputs_np = inputs_dask.compute()
        outputs_np = outputs_dask.compute()

        self.inputs = torch.from_numpy(inputs_np).float()
        self.outputs = torch.from_numpy(outputs_np).float()

        if torch.isnan(self.inputs).any() or torch.isnan(self.outputs).any():
            raise ValueError("NaNs found in dataset")

    def __len__(self):
        return self.size

    def __getitem__(self, idx):
        return self.inputs[idx], self.outputs[idx]

def get_lat_weights(latitude_values):
    lat_rad = np.deg2rad(latitude_values)
    weights = np.cos(lat_rad)
    return weights / np.mean(weights)
    
class ClimateDataModule(pl.LightningDataModule):
    def __init__(
        self,
        path,
        input_vars,
        output_vars,
        train_ssps,
        test_ssp,
        target_member_id,
        val_split=0.1,
        test_months=120,
        batch_size=32,
        num_workers=0,
        seed=42,
    ):
        super().__init__()
        self.path = path
        self.input_vars = input_vars
        self.output_vars = output_vars
        self.train_ssps = train_ssps
        self.test_ssp = test_ssp
        self.target_member_id = target_member_id
        self.val_split = val_split
        self.test_months = test_months
        self.batch_size = batch_size
        self.num_workers = num_workers
        self.seed = seed
        self.normalizer = Normalizer()

    def prepare_data(self):
        assert os.path.exists(self.path), f"Data path not found: {self.path}"

    def setup(self, stage=None):
        ds = xr.open_zarr(self.path, consolidated=False, chunks={"time": 24})
        spatial_template = ds["rsdt"].isel(time=0, ssp=0, drop=True)

        def load_ssp(ssp):
            input_dask, output_dask = [], []
            for var in self.input_vars:
                da_var = ds[var].sel(ssp=ssp)
                if "latitude" in da_var.dims:
                    da_var = da_var.rename({"latitude": "y", "longitude": "x"})
                if "member_id" in da_var.dims:
                    da_var = da_var.sel(member_id=self.target_member_id)
                if set(da_var.dims) == {"time"}:
                    da_var = da_var.broadcast_like(spatial_template).transpose("time", "y", "x")
                input_dask.append(da_var.data)

            for var in self.output_vars:
                da_out = ds[var].sel(ssp=ssp, member_id=self.target_member_id)
                if "latitude" in da_out.dims:
                    da_out = da_out.rename({"latitude": "y", "longitude": "x"})
                output_dask.append(da_out.data)

            return da.stack(input_dask, axis=1), da.stack(output_dask, axis=1)

        train_input, train_output, val_input, val_output = [], [], None, None

        for ssp in self.train_ssps:
            x, y = load_ssp(ssp)
            if ssp == "ssp370":
                val_input = x[-self.test_months:]
                val_output = y[-self.test_months:]
                train_input.append(x[:-self.test_months])
                train_output.append(y[:-self.test_months])
            else:
                train_input.append(x)
                train_output.append(y)

        train_input = da.concatenate(train_input, axis=0)
        train_output = da.concatenate(train_output, axis=0)

        self.normalizer.set_input_statistics(
            mean=da.nanmean(train_input, axis=(0, 2, 3), keepdims=True).compute(),
            std=da.nanstd(train_input, axis=(0, 2, 3), keepdims=True).compute(),
        )
        self.normalizer.set_output_statistics(
            mean=da.nanmean(train_output, axis=(0, 2, 3), keepdims=True).compute(),
            std=da.nanstd(train_output, axis=(0, 2, 3), keepdims=True).compute(),
        )

        train_input_norm = self.normalizer.normalize(train_input, "input")
        train_output_norm = self.normalizer.normalize(train_output, "output")
        val_input_norm = self.normalizer.normalize(val_input, "input")
        val_output_norm = self.normalizer.normalize(val_output, "output")

        test_input, test_output = load_ssp(self.test_ssp)
        test_input = test_input[-self.test_months:]
        test_output = test_output[-self.test_months:]
        test_input_norm = self.normalizer.normalize(test_input, "input")

        self.train_dataset = ClimateDataset(train_input_norm, train_output_norm)
        self.val_dataset = ClimateDataset(val_input_norm, val_output_norm)
        self.test_dataset = ClimateDataset(test_input_norm, test_output, output_is_normalized=False)

        self.lat = spatial_template.y.values
        self.lon = spatial_template.x.values
        self.area_weights = xr.DataArray(get_lat_weights(self.lat), dims=["y"], coords={"y": self.lat})

    def train_dataloader(self):
        return DataLoader(self.train_dataset, batch_size=self.batch_size, shuffle=True,
                          num_workers=self.num_workers, pin_memory=True)

    def val_dataloader(self):
        return DataLoader(self.val_dataset, batch_size=self.batch_size, shuffle=False,
                          num_workers=self.num_workers, pin_memory=True)

    def test_dataloader(self):
        return DataLoader(self.test_dataset, batch_size=self.batch_size, shuffle=False,
                          num_workers=self.num_workers, pin_memory=True)

    def get_lat_weights(self):
        return self.area_weights

    def get_coords(self):
        return self.lat, self.lon
        
class Normalizer:
    def __init__(self):
        self.mean_in, self.std_in = None, None
        self.mean_out, self.std_out = None, None

    def set_input_statistics(self, mean, std):
        self.mean_in = mean
        self.std_in = std

    def set_output_statistics(self, mean, std):
        self.mean_out = mean
        self.std_out = std

    def normalize(self, data, data_type):
        if data_type == "input":
            return (data - self.mean_in) / self.std_in
        elif data_type == "output":
            return (data - self.mean_out) / self.std_out

    def inverse_transform_output(self, data):
        return data * self.std_out + self.mean_out
````

## ðŸ“Š Exploratory Plots for Each Variable

In this section, we generate a series of visualizations for every input and output variable in the dataset to better understand their characteristics and behavior:

- **Histograms** (`*_hist.png`) show the overall value distribution of each variable across all timesteps and spatial grid cells.
- **Global temporal trends** (`*_temporal.png`) display the global mean evolution of each variable over time.
- **Scenario-specific trends** (`*_ssp_trend.png`) help compare how different SSPs (Shared Socioeconomic Pathways) influence the temporal behavior of a variable.
- **Spatial maps** (`*_spatial_mean.png`, `*_spatial_std.png`) illustrate how each variableâ€™s average and variability is distributed geographically across the globe.

These visualizations are crucial for identifying non-stationarity, scenario effects, and spatial structures in the climate variables, which can guide model design and preprocessing choices.


````python
def plot_all_histograms_cleaned(ds, vars_list, out_dir, n_cols=2, clip_percentiles=(1, 99)):
    """
    Plot histograms for all variables in a stacked layout with clipping and log-scale handling.
    """
    hist_data = []
    for v in vars_list:
        raw = ds[v].values
        finite = raw[np.isfinite(raw)].reshape(-1)
        if len(finite) == 0:
            continue
        hist_data.append((v, finite))

    n_vars = len(hist_data)
    n_rows = int(np.ceil(n_vars / n_cols))
    fig, axes = plt.subplots(n_rows, n_cols, figsize=(5 * n_cols, 3.5 * n_rows))
    axes = axes.flatten()

    for idx, (var, arr) in enumerate(hist_data):
        ax = axes[idx]
        # clip outliers to 1stâ€“99th percentile range
        vmin, vmax = np.percentile(arr, clip_percentiles)
        clipped = arr[(arr >= vmin) & (arr <= vmax)]

        # determine if log scale is helpful
        use_log = (clipped > 0).all() and np.log10(clipped.max()) - np.log10(clipped.min()) > 3

        ax.hist(np.log10(clipped) if use_log else clipped, bins=80, density=True, alpha=0.8)
        ax.set_title(var, fontsize=10)
        ax.set_xlabel("log10(Value)" if use_log else "Value", fontsize=8)
        ax.set_ylabel("Density", fontsize=8)
        ax.tick_params(axis="both", labelsize=7)

    for i in range(len(hist_data), len(axes)):
        axes[i].axis("off")

    plt.tight_layout()
    plt.savefig(out_dir / "histograms_all_fixed.png")
    plt.show()
    plt.close()


plot_all_histograms_cleaned(ds, all_vars, out_dir, n_cols=3, use_log_y=False, use_density=True)
````

````python
def save_hist(arr, name):
    """Save histogram plot of array values."""
    arr = arr[np.isfinite(arr)]  # filter out NaN and inf
    if len(arr) == 0:
        print(f"[skip] {name}: no finite values to plot")
        return
    plt.figure(figsize=(4, 3))
    plt.hist(arr, bins=100, density=True)
    plt.title(f"Histogram of {name}")
    plt.tight_layout()
    plt.savefig(out_dir / f"{name}_hist.png")
    plt.close()

def save_temporal(var, by_ssp=False):
    """
    Plot global-mean value over time (optionally by SSP).
    Uses (year + month / 12) to avoid datetime plotting issues.
    """
    plt.figure(figsize=(10, 3))
    if by_ssp and "ssp" in ds[var].dims:
        for s in ds.ssp.values:
            ser = ds[var].sel(ssp=s).mean([d for d in ds[var].dims if d not in {"time", "ssp"}])
            x = ser["time"].dt.year + (ser["time"].dt.month - 1) / 12
            plt.plot(x, ser, label=s)
        plt.legend()
        suffix = "_ssp_trend"
    else:
        ser = ds[var].mean([d for d in ds[var].dims if d != "time"])
        x = ser["time"].dt.year + (ser["time"].dt.month - 1) / 12
        plt.plot(x, ser)
        suffix = "_temporal"

    plt.title(f"Global mean {var} over time")
    plt.xlabel("year")
    plt.ylabel(var)
    plt.tight_layout()
    plt.savefig(out_dir / f"{var}{suffix}.png")
    plt.close()

def save_spatial(var, stat="mean"):
    """Plot time-mean or time-std of a variable as a 2D spatial map (if applicable)."""
    # Step 1: Compute time statistic
    field = ds[var].mean("time") if stat == "mean" else ds[var].std("time")

    # Step 2: Drop or average over any remaining non-spatial dims (e.g., 'ssp')
    dims_to_reduce = [d for d in field.dims if d not in ("x", "lon", "longitude", "y", "lat", "latitude")]
    if dims_to_reduce:
        field = field.mean(dims_to_reduce)

    # Step 3: Detect spatial dimensions
    x_name = next((d for d in field.dims if d in ("x", "lon", "longitude")), None)
    y_name = next((d for d in field.dims if d in ("y", "lat", "latitude")), None)

    if x_name is None or y_name is None:
        print(f"[skip] {var}: unable to find spatial dims â€” skipping")
        return

    # Step 4: Plot
    plt.figure(figsize=(6, 3))
    im = plt.pcolormesh(field[x_name], field[y_name], field, shading="auto")
    plt.colorbar(im, label=f"{var} ({stat})")
    plt.title(f"Time-{stat} {var}")
    plt.tight_layout()
    plt.savefig(out_dir / f"{var}_spatial_{stat}.png")
    plt.show()
    plt.close()

# ------------------------------------------------------------------
# 5. Plots for every variable
# ------------------------------------------------------------------
for v in all_vars:
    save_hist(ds[v].values.reshape(-1), v)
    save_temporal(v)
    save_temporal(v, by_ssp=True)
    save_spatial(v, "mean")
    save_spatial(v, "std")
````

# Heatmap

````python
# Filter variables with matching total number of elements
flattened_vars = []
valid_var_names = []
reference_shape = None

for v in all_vars:
    try:
        arr = ds[v].values.reshape(-1)
        if np.all(np.isfinite(arr)):  # skip if all NaNs
            if reference_shape is None:
                reference_shape = arr.shape
            if arr.shape == reference_shape:
                flattened_vars.append(arr)
                valid_var_names.append(v)
            else:
                print(f"[skip] {v}: shape {arr.shape} does not match reference {reference_shape}")
        else:
            print(f"[skip] {v}: contains non-finite values")
    except Exception as e:
        print(f"[skip] {v}: error {e}")

if len(flattened_vars) >= 2:
    corr = np.corrcoef(flattened_vars)
    plt.figure(figsize=(6, 5))
    sns.heatmap(corr, annot=True, fmt=".2f",
                xticklabels=valid_var_names, yticklabels=valid_var_names,
                cmap="coolwarm", vmin=-1, vmax=1)
    plt.title("Pearson correlation across variables")
    plt.tight_layout()
    plt.show()
else:
    print("Not enough valid variables for correlation heatmap.")
````
